```python
def json_lst_in(lst_input):
    out = {}
    
    def flat(lst_input, name = ""):
        i = 0
        
        for a in lst_input:
            flat(a, name + str(i) + "_")
            i += 1
            
    flat(lst_input)
    return out
```
